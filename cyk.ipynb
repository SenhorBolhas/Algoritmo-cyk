```python
### Leitura do arquivo de texto que contém a gramática
arquivo = open('gramatica.txt','r')
    if arquivo.mode == 'r':
        gramatica = arquivo.read()
        arquivo.close()
```

```python
### Derivação
def derivacao(array1, array2, dicionario):
    resultado = []
    if len(array1) == 0:
        for o2 in array2:
            if o2 in dicionario.keys():
                for a in dicionario[o2]:
                    resultado.append(o)
    elif len(array2) == 0:
        for o1 in array1:
            if o1 in dicionario.keys():
                for a in dicionario[o1]:
                    resultado.append(o)
    else:
        for o1 in array1:
            for o2 in array2:
                aux = o1 + o2
                if aux in dicionario.keys():
                    for o in dicionario[aux]:
                        resultado.append(o)
    return resultado
```

```python
### Registrando a gramática lida do arquivo
def get_gramatica(gramatica):
    aux = gramatica.split('\n')
    esquerda = []
    direita = []
    for i in aux:
        aux1 = i.split(' => ')
        esquerda.append(aux1[0])
        direita.append(aux1[1])
    aux1 = []
    for i in range(len(direita)):
        aux1.append(direita[i].split(' | '))
    dicionario = {}
    for i in range(len(aux1)):
        for k in aux1[i]:
            if k in dicionario.keys():
                if esquerda[i] not in dicionario[k]:
                    dicionario[k].append(esquerda[i])
            else:
                dicionario[k] = [esquerda[i]]
    return dicionario
```

```python
### Executando o cyk
def cyk(palavra, gramatica=gramatica):
    
    dicionario = get_gramatica(gramatica)
    
    arvore = []
    linha = []
    for car in palavra:
        line.append(dicionario[car])
    arvore.append(linha)

    for h in range(1,len(palavra)):
        linha = []
        for x in range(0,len(palavra) - h):
            subline = []
            for y in range(h):
                for d in derivacao(arvore[y][x],arvore[h - y - 1][x + 1 + y],dicionario):
                    subline.append(d)
            linha.append(subline)
        arvore.append(linha)
    
    return arvore
```
